# Dados E-commerce

```python
import pandas as pd
```

```python
df = pd.read_csv('dados_ecommerce.csv')
df.head()
```

```python
df.info()
```

```python
# Trocar s/n para nan
import numpy as np

df['Avaliação'] = df['Avaliação'].replace('s/n', np.nan)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Preço'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Categoria'] = df['Categoria'].fillna("Sem Categoria")
```

```python
df['Categoria'].unique()
```

```python
df.loc[df['Categoria'] == 'eletronicos', 'Categoria'] = 'Eletrônicos'
```

```python
df['Categoria'].unique()
```

```python
df.isnull().sum()
```

```python
# converter o tipo
df['Avaliação'] = pd.to_numeric(df['Avaliação'])
```

```python
# media
df['Avaliação_media'] = df['Avaliação']
media_avaliacao = df['Avaliação_media'].mean()
df['Avaliação_media'] = df['Avaliação_media'].fillna(media_avaliacao)
```

```python
df['Avaliação_media_cat'] = df['Avaliação']
media_cat = df.groupby('Categoria')['Avaliação'].mean()
```

```python
df['Avaliação_media_cat'] = df['Avaliação_media_cat'].fillna(df['Categoria'].map(media_cat))
```

```python
df
```

```python
df['Produto'] = df['Produto'].str.lower().str.replace(' ', '')

df
```

```python
df[df['Quantidade'] < 0]
df.loc[df['Quantidade'] < 0, 'Quantidade'] = 0

# jeito alternativo - igual
df['Quantidade'] = df['Quantidade'].clip(lower=0)
```

```python
df
```

```python
df['Avaliação'] = df['Avaliação'].clip(lower=1, upper=5)
#mudar para todas as colunas de avaliação
```

```python
df
```

```python
df[df.duplicated(subset=['Produto'])]
```

```python
df.drop_duplicates(subset=['Produto'], keep='first', inplace=True)
```
